```python
#to plot a linear graph
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
y=[2,4,1,5,3,8]
plt.plot(x,y)
plt.show()
```

```python
#Sine wave plotting
import numpy as np
import matplotlib.pyplot as plt
t=np.arange(0.0,2.0,0.01)
s=np.sin(2*np.pi*t)
plt.plot(t,s)
plt.xlabel('t')
plt.ylabel('sin(x)')
plt.title('sine wave plot')
plt.show()
```

```python
#Cosine wave plotting
import numpy as np
import matplotlib.pyplot as plt
t=np.arange(0.0,2.0,0.01)
s=np.cos(2*np.pi*t)
plt.plot(t,s)
plt.xlabel('t')
plt.ylabel('cos(x)')
plt.title('cosine wave plot')
plt.show()
```

```python
#Line Sampling
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
y=[2,4,1,5,6,2]
plt.plot(x,y,color="blue",linestyle="dashed",linewidth=4,marker='o',markerfacecolor="orange",markersize=12)
plt.xlabel('x')
plt.ylabel('y')
plt.title('LINE SAMPLING')
plt.show()
```

```python
#BAR GRAPH

import numpy as np
import matplotlib.pyplot as plt

objects = ('Python', 'C++', 'Java', 'Perl', 'Scala', 'Lisp')
y = np.arange(len(objects))
performance = [10,8,6,4,2,1]

plt.bar(y, performance, align='center', alpha=1.0)
plt.xticks(y, objects)
plt.ylabel('Usage')
plt.title('Programming language usage')

plt.show()
```

```python
#BAR GRAPH

import numpy as np
import matplotlib.pyplot as plt

objects = ('Python', 'C++', 'Java', 'Perl', 'Scala', 'Lisp')
y = np.arange(len(objects))
performance = [10,8,6,4,2,1]

plt.barh(y, performance, align='center', alpha=1.0)
plt.yticks(y, objects)
plt.xlabel('Usage')
plt.title('Programming language usage')

plt.show()
```

```python
#BAR GRAPH

import numpy as np
import matplotlib.pyplot as plt

l=[1,2,3,4,5]
h=[10,24,36,45,10]
label=['JAN','FEB','MARCH','APRIL','MAY']
plt.bar(l,h,tick_label=label,width=0.8,color=['green','blue'])
plt.xlabel('left')
plt.ylabel('height')
plt.title('BAR GRAPH')
plt.show()
```

```python
#HISTOGRAM

import numpy as np
import matplotlib.pyplot as plt
x=[12,56,34,78,6,54,22,45,14]
plt.hist(x,color='red',histtype='bar',rwidth=0.9)
plt.xlabel('X')
plt.ylabel('Y')
plt.title("HISTOGRAM")
plt.show()
```

```python
#SCATTER PLOT

import numpy as np
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[1,4,9,16,25]
plt.scatter(x,y,label='diamond',color='blue',marker='D')
plt.xlabel('X')
plt.ylabel('Y')
plt.title("scatter plot")
plt.show()
```

```python
#pieChart

import matplotlib.pyplot as plt

# Data to plot
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]
colors = ['blue', 'yellow', 'green', 'red']
explode = (0.1, 0, 0, 0)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140)

plt.axis('equal')
plt.show()
```
